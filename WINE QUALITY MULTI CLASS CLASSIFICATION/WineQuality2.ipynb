```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine # import wine quality Dataset
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# 1. Problem Statement & Data Loading

data_ar = pd.read_csv('/content/WineQT.csv')


# 2. Exploratory Data Analysis (EDA) & Data Visualization
print(f"Dataset shape:{data_ar.shape}")
print("Data Information:")
print(data_ar.info())
print("\nDescriptive Statistics:")
print(data_ar.describe())
print("\nQuality Distribution:")  # wine quality counts
print(data_ar["quality"].value_counts())
```

```python
# Visualizations of Quality


print("\nVisualizations : Histograms")
for colm in data_ar:
    plt.figure(figsize=(8, 6))
    sns.histplot(data_ar[colm], kde=True)
    plt.title(f'Distribution of {colm}')
    plt.show()

    # Box Plot
print("\nVisualizations of target: BoxPloats")
for colm in data_ar:
    plt.figure(figsize=(8, 6))

    sns.boxplot(y=data_ar[colm], color='lightgreen')

    plt.title(f'Box Plot of {colm}')

    plt.show()

 # Density Plot
print("\nVisualizations of target: Density Ploats")
for colm in data_ar:
    plt.figure(figsize=(8, 6))

    sns.kdeplot(data_ar[colm],fill=True, color='salmon')

    plt.title(f'Density Plot of {colm}')

    plt.tight_layout()
    plt.show()

#Scatter plots

print("\nVisualizations of target: Scatter plots")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol',y='sulphates',hue='quality',data=data_ar)
plt.title( 'Scatter plot: alcohol  Vs sulphates')
plt.show()

#Pair plots
print("\nVisualizations of target: Pair plots")
sns.pairplot(data_ar, hue='quality')
plt.show()


#Count Ploats
print("\nVisualizations of target: Count Ploats")
sns.countplot(x='quality', data=data_ar)
plt.title('Distribution of Wine Quality Labels')
plt.show()


#bar charts
print("\nVisualizations of target: Bar Chart")
sns.countplot(x='quality', data=data_ar,color='red')
plt.title('Distribution of Wine Quality Labels')
plt.show()
```
